## Question 6a

Plot the waveform
	 $$x(t) = e^{-t^2}\cos(2\pi t)$$
	 for $-5\leq t\leq 5$, with a step size of 0.01.

```python
step_size = 0.01 # YOUR CODE HERE: enter the appropriate step size here
t = np.arange(-5, 5, step_size)

x = np.exp(-t**2) * np.cos(2*np.pi*t) # YOUR CODE HERE: enter the function above here
plt.figure()
plt.plot(t, x)
plt.show()
```

## Question 6b

Create a function relu(t) that implements the relu function from Question 1,\
Then plot the function for $-5\leq t\leq 5$, with a step size of 0.01.

```python
step_size = 0.01 # YOUR CODE HERE: enter the appropriate step size here
t = np.arange(-5, 5, step_size)
def relu(t):
    # YOUR CODE HERE: enter the relu function from Question 1 here
    out_signal = []
    for val in t:
        if val >= 0:
            out_signal.append(val)
        else:
            out_signal.append(0)
    return out_signal

plt.figure()
plt.plot(t, relu(t))
plt.show()
```
