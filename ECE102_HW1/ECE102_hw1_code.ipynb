## Question 6a

Plot the waveform
	 $$x(t) = e^{-t^2}\cos(2\pi t)$$
	 for $-5\leq t\leq 5$, with a step size of 0.01.

```python
step_size = 0.01 # YOUR CODE HERE: enter the appropriate step size here
t = np.arange(-5, 5, step_size)

x = np.exp(-t**2) * np.cos(2 * np.pi * t) # YOUR CODE HERE: enter the function above here
plt.figure()
plt.plot(t, x)
plt.show()
```

## Question 6b

Create a function relu(t) that implements the relu function from Question 1,\
Then plot the function for $-5\leq t\leq 5$, with a step size of 0.01.

```python
step_size = 0.01 # YOUR CODE HERE: enter the appropriate step size here
t = np.arange(-5, 5, step_size)
def relu(t):
    # YOUR CODE HERE: enter the relu function from Question 1 here
    return np.maximum(0, t)
plt.figure()
plt.plot(t, relu(t))
plt.show()
```

## Question 6c

Create functions even(t, f) and odd(t, f) that take inputs time t and function f that compute the \
respective even and odd parts of f(t) at points t.

```python
# Here is an example of a function that takes in another function as input
def square(t, f):
    return np.power(f(t), 2)
plt.figure()
plt.plot(t, square(t, relu))
plt.show()
```

```python
# Do this for the following functions even and odd:
def even(t, f):
    # YOUR CODE HERE
    return (f(t) + f(-t)) / 2

def odd(t, f):
    # YOUR CODE HERE
    return (f(t) - f(-t)) / 2

# Plotting code
plt.figure()
plt.plot(t, even(t, relu))
plt.show()

plt.figure()
plt.plot(t, odd(t, relu))
plt.show()
```
