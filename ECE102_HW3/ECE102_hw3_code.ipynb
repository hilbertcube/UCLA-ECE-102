```python
import numpy as np
import matplotlib.pyplot as plt
```

For this homework question, we will be computing the convolution of two rect signals to see what the resulting output looks like. In the code blocks below we will be creating two rect signals, and using python numpy library functions to perform the convolution. Hint: check out the numpy.convolve() function documentation.

(2 points) In the cell below, we will start by making the rect function.

```python
# Make the x-axis first to use when constructing the rect signal
x = np.linspace(-1, 1, 100)

rect = np.where(np.abs(x) <= 0.5, 1, 0)

# Plotting code
plt.plot(x, rect)
plt.title('Rectangular Function')
plt.xlabel('x')
plt.ylabel('rect(x)')
plt.grid(True)
plt.show()
```

(8 points) In the cell below, you will use what you wrote in the previous part to setup two rect signals and convolve them using numpy convolve. We will then graph it and see the signal output

```python
# Make the x-axis first to use when constructing the rect signal
x = np.linspace(-1, 1, 100)

# Create two rect functions
rect1 = np.where(np.abs(x) <= 0.5, 1, 0)
rect2 = np.where(np.abs(x) <= 0.5, 1, 0)

out = np.convolve(rect1, rect2, mode='full')

# Plotting code
x_plot = np.linspace(-2, 2, out.shape[0])
plt.plot(x_plot, out)
plt.title('Convolution of Two Rectangular Functions')
plt.xlabel('x')
plt.ylabel('rect(x) * rect(x)')
plt.grid(True)
plt.show()
```

What does this output look like? Do you have any guesses as to why? (No need to respond to this last question, just an optional thing to think about if interested)

```python
# Make the x-axis first to use when constructing the rect signal
x = np.linspace(-1, 1, 200)

# Create the initial rect function
rect = np.where(np.abs(x) <= 0.5, 1, 0)

# Start with the first rect function
rect_out = rect.copy().astype(float)

# Convolve 10 rect functions together
for i in range(9):
    rect_out = np.convolve(rect_out, rect, mode='full')
    # Normalize to prevent numerical overflow
    rect_out = rect_out / np.sum(rect)

# Plotting code
x_plot = np.linspace(-5, 5, rect_out.shape[0])
plt.figure(figsize=(10, 6))
plt.plot(x_plot, rect_out)
plt.title('Convolution of 10 Rectangular Functions (Normalized)')
plt.xlabel('x')
plt.ylabel('Normalized Amplitude')
plt.grid(True)
plt.show()
```
